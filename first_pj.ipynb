```python
import pandas as pd
import sqlite3
import os

os.getcwd()
```

```python
#Created the sqlite DataBase
db = sqlite3.connect("PythonDataBase.db")

#read CSV into datafrome
df3 = pd.read_csv("trips.csv", header="infer")

#convert the pandas datatframe to a sqlite table
df3.to_sql("CashOut", db, if_exists="replace")

def run_query(query):
    return pd.read_sql_query(query,db)
```
